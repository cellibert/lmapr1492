On nous demande de déterminer les 3 premiers pics du diffractogramme correspondant (en indiquant les indices hkl associés) pour un rayonnement X incident Cu Kα de longueur d’onde λ=1.54060 Å.

```python
from pymatgen.ext.matproj import MPRester
from pymatgen.analysis.diffraction import xrd
from matplotlib import *
```

```python
assigned_ID = "mp-7173"
chemical_formula = "ScSbPt"

# ----------==========     Procédure de téléchargement du fichier .CIF assigné     ==========----------

with MPRester("L41ZHCirMyRASy8QHJEO") as myCristal:
    
    structure = myCristal.get_structure_by_material_id(assigned_ID)
```

```python
# ----------==========     Calcul des pics du diffractogramme demandés     ==========----------
wave_length = 1.54060
nbr_pics = 3

myXRay        = xrd.XRDCalculator(wavelength=wave_length,symprec=0, debye_waller_factors=None)
diffractogram = myXRay.get_pattern(structure,scaled=True,two_theta_range=(0,90))

hkl         = diffractogram.hkls
angles      = diffractogram.x
intensities = diffractogram.y
```

```python
# ----------==========     Affichage des résultats demandés     ==========----------

print("Determination des 3 premiers pics du diffractogramme correspondant") 
print("pour un rayonnement X incident Cu Kα (\u03BB = 1.54060 \u212B)")
print("avec affichage des indices hkl associes.\n")

for i in range(nbr_pics):
    print(" -Pic Nr."+str(i+1)+"- \n")
    print(" > Angle : 2\u03B8   = " + str(angles[i]))
    print(" > Intensite         = " + str(intensities[i]))
    print(" > Indices de Miller = " + str(hkl[i][0]['hkl']))
    print(" _____________________\n")

myXRay.show_plot(structure)
```
